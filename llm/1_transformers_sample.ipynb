```python
from transformers import pipeline
```

```python
# Document Classification (文書分類) : 何らかのグループに分類するタスク

text_classification_pipeline = pipeline(
    model ="llm-book/bert-base-japanese-v3-marc_ja" # 日本語の感情分析モデル
)

positive_text = "あのグループの音楽はとてもすばらしい"
negative_text = "あのグループの音楽は聴くに堪えない"

print(text_classification_pipeline(positive_text)[0])
print(text_classification_pipeline(negative_text)[0])
```

```python
# Natural Langage Inference (NLI;自然言語推論): 論理関係を予測する

nli_pipeline = pipeline(
    model = "llm-book/bert-base-japanese-v3-jnli"
)

text1 = "犬が吠えている"
text2 = "吠えている犬がいる"
text3 = "犬が走っている"
text4 = "犬が空に向かって吠えていた"

print(nli_pipeline({"text": text1, "text_pair": text2}))
print(nli_pipeline({"text": text1, "text_pair": text3}))
print(nli_pipeline({"text": text1, "text_pair": text4}))
```

```python
# Semantic Textual Similarity (意味的類似度計算): 類似度をスコアリングする

sts_pipeline = pipeline(
    model = "llm-book/bert-base-japanese-v3-jsts" ,
    function_to_apply = "none"
)

text1 = "犬が吠えている"
text2 = "吠えている犬がいる"
text3 = "犬が走っている"

print(sts_pipeline({"text": text1, "text_pair": text2})["score"])
print(sts_pipeline({"text": text1, "text_pair": text3})["score"])
```

```python
# Named Entity Recognition (NER;固有表現認識): 固有表現を抽出する

ner_pipeline = pipeline(
    model = "llm-book/bert-base-japanese-v3-ner-wikipedia-dataset" ,
    aggregation_strategy = "simple"
)

text = "江戸幕府の初代征夷大将軍は、徳川家康公である"

print(ner_pipeline(text))
```
